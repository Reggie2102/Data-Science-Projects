```python
import pandas as pd
import numpy as np
from google.colab import files
```

```python
uploaded = files.upload()
file_name = next(iter(uploaded))  # Gets the first uploaded filename
print(f"Uploaded: {file_name}")

# Load the dataset
df = pd.read_csv(file_name)
print(f"Shape: {df.shape}")
df.head()
```

```python
print("=== Data Types ===")
print(df.dtypes)

print("\n=== Missing Values ===")
print(df.isnull().sum())

# Quick stats
df.describe(include='all')
```

```python
print("Current missing values:")
print(df.isnull().sum())
```

```python
fill_rules = {
    # Numeric columns (40-260)
    **{str(col): df[str(col)].median() for col in range(40, 261, 10)},

    # Categorical/time column
    'time': 'Unknown_time'
}

# Apply filling (would change nothing in your current data)
df_filled = df.fillna(fill_rules)

# Verify (will show all 0s for your data)
print("\nMissing values after filling:")
print(df_filled.isnull().sum())

# =============================================
# 4. OUTLIER HANDLING (more relevant for your data)
# =============================================
# Your data has outliers (visible in the min/max stats)
for col in [str(x) for x in range(40, 261, 10)]:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    df[col] = df[col].clip(
        lower=q1 - 3*iqr,  # Conservative multiplier
        upper=q3 + 3*iqr
    )
```

```python
print(f"Before: {df.shape}")
df = df.drop_duplicates()
print(f"After: {df.shape}")
```

```python
output_file = 'cleaned_data.csv'
df.to_csv(output_file, index=False)

# Download to local machine
files.download(output_file)
```
