```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Set visual style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
%matplotlib inline
```

```python
# Upload Dataset
uploaded = files.upload()
file_name = next(iter(uploaded))
df = pd.read_csv(file_name)  # For Excel: pd.read_excel(file_name)
```

```python
print(f" Dataset loaded: {file_name}")
print(f"Shape: {df.shape} rows, columns")

print("\n First 5 rows:")
display(df.head())

print("\n Dataset Info:")
df.info()  # Shows data types and missing values

print("\n Basic Statistics:")
display(df.describe(include='all'))
```

```python
# Melt the dataframe
df_melted = df.melt(id_vars='time', var_name='time_point', value_name='expression')

# Boxplot
plt.figure(figsize=(15, 6))
sns.boxplot(data=df_melted, x='time_point', y='expression')
plt.xticks(rotation=45)
plt.title("Expression Distribution Across Time Points")
plt.show()
```

```python
# Mean expression per time point
mean_expr = df.drop(columns='time').mean()
mean_expr.plot(kind='line', marker='o')
plt.title("Average Gene Expression Over Time")
plt.xlabel("Time Point")
plt.ylabel("Expression Level")
plt.grid()
plt.show()
```

```python

from sklearn.preprocessing import StandardScaler
top_genes = df.set_index('time').var(axis=1).nlargest(50).index
subset = df[df['time'].isin(top_genes)].set_index('time')

# Normalize and plot
sns.clustermap(StandardScaler().fit_transform(subset), cmap='viridis')
plt.title("Clustermap of Top 50 Variable Genes")
plt.show()
```

```python
sns.heatmap(df.drop(columns='time').corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Between Time Points")
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df.drop(columns='time'))

plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.5)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("PCA of Gene Expression Across Time")
plt.show()
```
