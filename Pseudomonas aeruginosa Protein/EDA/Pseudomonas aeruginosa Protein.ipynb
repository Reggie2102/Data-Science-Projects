```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Set visual style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
%matplotlib inline
```

```python
# Upload Your Dataset
uploaded = files.upload()
file_name = next(iter(uploaded))
df = pd.read_csv(file_name)
```

```python
print(f" Dataset loaded: {file_name}")
print(f"Shape: {df.shape} rows, columns")


print("\n🔍 First 5 rows:")
display(df.head())

print("\nℹ️ Dataset Info:")
df.info()

print("\n📊 Basic Statistics:")
display(df.describe(include='all'))
```

```python
#Molecular Weight Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Molecular_Weight', bins=50)
plt.title("Distribution of Molecular Weights (Da)")
plt.axvline(df['Molecular_Weight'].median(), color='red', linestyle='--', label='Median')
plt.show()
```

```python
#Isoelectric Point vs Hydrophobicity
sns.jointplot(data=df, x='Isoelectric_Point', y='Hydrophobicity', kind='hex')
plt.suptitle("pI vs Hydrophobicity Relationship")
plt.tight_layout()
plt.show()
```

```python
#Amino Acid Composition Analysis
# Convert dictionary column to DataFrame
import ast
aa_df = pd.json_normalize(df['Amino_Acid_Composition'].apply(ast.literal_eval)).fillna(0)
top_aas = aa_df.sum().sort_values(ascending=False).head(10)

top_aas.plot(kind='bar')
plt.title("Top 10 Most Frequent Amino Acids")
plt.ylabel("Total Count Across All Proteins")
plt.show()
```

```python
long_seqs = df.nlargest(5, 'Seq_Length')[['ID', 'Name', 'Seq_Length']]
print("Longest sequences:\n", long_seqs)
```

```python
#MW Calculation Discrepancy
df['MW_diff'] = df['Molecular_Weight'] - df['MW_Calc']
sns.boxplot(data=df, y='MW_diff')
plt.title("Discrepancy Between Reported and Calculated MW")
plt.show()
```
